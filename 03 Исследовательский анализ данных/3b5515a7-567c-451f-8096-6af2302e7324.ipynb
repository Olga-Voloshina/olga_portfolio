### Заполнение пропусков

При изучении файла мы обнаружили, что многие столбцы содержат пропуски, некоторые столбцы заполнены на 99%, другие не более чем на 10%. Рассмотрим каждый столбец с пропусками отдельно, заполним пропуски подходящими значениями.

Первый по порядку столбец с пропусками - 'ceiling_height'. Предположим, что высота потолка не указана без каких-либо причин и заполним пропуски медианным значением.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Далее столбец 'floors_total'. Пропущеных значений мало, на статистику они повлияют слабо. Заполним пропуски значениями из столбца 'floor', как минимальными возможными. Следует помнить, что заполнение такими значениями приведет к искусственному увеличению доли квартир, расположенных на верхнем этаже. В данном случае количество пропусков незначительно, и изменение доли квартир на верхних этажах будет невелико.

```python
data['floors_total'] = data['floors_total'].fillna(data['floor'])
```

Следующие столбцы - 'living_area'  и 'kitchen_area'. Заполним пропуски, взяв за основу данные из столбца 'total_area', помножим на медианное отношение жилой площади к общей, или, соответственно, площади кухни к общей.

```python
living_coef = (data['living_area']/data['total_area']).median()
kitchen_coef = (data['kitchen_area']/data['total_area']).median()
data['living_area'] = data['living_area'].fillna(data['total_area']*living_coef)
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*kitchen_coef)
```

Добавим столбец с характеристикой этажа квартиры: первый, последний, другой. Жилье в одноэтажных строениях будем считать за первый этаж. Будем иметь ввиду, что пропущенные значения в колонке 'floors_total' заполняли значениями этажа квартиры, то есть около 70 значений 'последний' в новой колонке могут быть ошибочными. 

```python
def floor_type(row):
    if row['floor']==1:
        return 'первый'
    elif row['floor']==row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
data['floor_type'] = data.apply(floor_type, axis=1)
```

Рассчитаем соотношение жилой и общей площади, а также отношение площади кухни к общей. Запишем результат соответственно в два столбца. Это значит, что в новых столбцах по ~2.5 тысячи значений будут одинаковыми и равными медиане по сырым данным.

```python
data['living_ratio'] = data['living_area']/data['total_area']
data['kitchen_ratio'] = data['kitchen_area']/data['total_area']
```
